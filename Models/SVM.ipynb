```python
import pandas as pd
import numpy as np
```

```python
# Read the csv file into a pandas DataFrame
df = pd.read_csv('../Data/df_clinical_ML_filtered.csv')
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
# Set features. This will also be used as your x values.
selected_features = df.drop(['overall_survival'], axis=1)
selected_features
```

```python
# Set X and y variables
y = df['overall_survival']
X = selected_features

print(X.shape, y.shape)
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , random_state= 1)
```

```python
X_train.head()
```

```python
# Scale your data
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from tensorflow.keras.utils import to_categorical

# Transform the training and testing data using the X_scaler
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

# Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# One-hot encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train_scaled, encoded_y_train)
predictions = model.predict(X_test_scaled)
model
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, encoded_y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, encoded_y_test)}")
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(encoded_y_test, predictions, target_names= ['1', '0']))
```

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1,5,10,50],
                'gamma': [0.0001, 0.0005, 0.0001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose = 3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, encoded_y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
